```python
import whisper
from pathlib import Path
import glob


AUDIO_DIR = "/content/drive/MyDrive/dataPFA/meetings_audio_wav"
audio_files = sorted(glob.glob(f"{AUDIO_DIR}/*.wav"))

print(f"{len(audio_files)} fichiers audio trouvés 🎧")


```

```python
# 🔊 Charger le modèle multilingue
model = whisper.load_model("large")

# 🌀 Transcription automatique
for audio_path in audio_files:
    result = model.transcribe(audio_path, language=None)  # auto-détection
    print(f"\n🎙️ {Path(audio_path).name}")
    print(f"Langue détectée : {result['language']}")
    print("Texte :", result["text"])
```
